Importing matplotlib and read image from detecto to plot the image

```python
import matplotlib.pyplot as plt
from detecto.utils import read_image
```

```python
base_path = "/content/drive/MyDrive/Dataset/"
sample_image = "dd9019db-8f94-11ed-aef7-2811a849f5b5.jpg"
image = read_image(base_path + sample_image)
plt.imshow(image)
plt.show()
```

Importing Databse from detecto and show_labeled_image to detect the facemask with my labeled images

```python
from detecto.core import Dataset
dataset = Dataset(base_path)
```

```python
from detecto.visualize import show_labeled_image

image, targets = dataset[0]
show_labeled_image(image, targets['boxes'], targets['labels'])
```

```python
from detecto.visualize import show_labeled_image

image, targets = dataset[60]
show_labeled_image(image, targets['boxes'], targets['labels'])
```

Initiating my model and labels

```python
from detecto.core import DataLoader, Model
labels = ['Mask', 'NoMask']
model = Model(labels)
```

Training the model with my database

```python
model.fit(dataset)
```

Here is where the program read the image and using the model it predict the labels in order of probability.

```python
image = read_image(base_path + "dd9019db-8f94-11ed-aef7-2811a849f5b5.jpg")
labesls, boxes, scores = model.predict(image)
```

```python
print("labels", labels)
print("boxes", boxes)
print("score", scores)
```

Here I call show labeled image to show the image, the box in witch the mask is placed and the label.

```python
from detecto.visualize import show_labeled_image
show_labeled_image(image, boxes[0], labels[0])
```
